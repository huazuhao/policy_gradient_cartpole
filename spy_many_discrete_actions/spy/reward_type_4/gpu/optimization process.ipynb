```python
import jsonpickle
```

```python
optimization_process_file_name = 'optimization_history.json'
```

```python
optimization_history = jsonpickle.decode(open(optimization_process_file_name).read())
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
optimization_history = optimization_history['history']
```

```python
fig = plt.figure(figsize=(7, 5))
matplotlib.rcParams.update({'font.size': 16})
plt.plot(optimization_history, 'b.', ms=10)  # Plot all evaluated points as blue dots
plt.plot(np.maximum.accumulate(optimization_history), 'r', lw=3)  # Plot cumulative minimum as a red line

plt.xlim([0, len(optimization_history)])
#plt.ylim([0, np.max(optimization_history)])
plt.ylim([0, 500])
plt.title("vix trading model 5")

plt.tight_layout()
plt.show()
```
