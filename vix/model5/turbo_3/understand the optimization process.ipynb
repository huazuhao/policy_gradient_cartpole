```python
import jsonpickle
```

```python
optimization_process_file_name = 'optimization_process_data.json'
```

```python
fX_dict = jsonpickle.decode(open(optimization_process_file_name).read())
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
fX = fX_dict['fx']
```

```python
fig = plt.figure(figsize=(7, 5))
matplotlib.rcParams.update({'font.size': 16})
plt.plot(fX, 'b.', ms=10)  # Plot all evaluated points as blue dots
plt.plot(np.minimum.accumulate(fX), 'r', lw=3)  # Plot cumulative minimum as a red line
plt.xlim([0, len(fX)])
plt.ylim([np.min(fX), 100])
plt.title("vix trading model 5")

plt.tight_layout()
plt.show()
```

```python
optimization_process_file_name = 'optimization_history.json'
history_dict = jsonpickle.decode(open(optimization_process_file_name).read())
```

```python
history = history_dict['history']
```

```python
plt.plot(history[1000:])
plt.show()
```

```python
fig = plt.figure(figsize=(7, 5))
matplotlib.rcParams.update({'font.size': 16})
plt.plot(history, 'b.', ms=10)  # Plot all evaluated points as blue dots
plt.plot(np.maximum.accumulate(history), 'r', lw=3)  # Plot cumulative minimum as a red line
plt.xlim([0, len(history)])
plt.ylim([-100, np.max(history)])
plt.title("vix trading model 5")

plt.tight_layout()
plt.show()
```
