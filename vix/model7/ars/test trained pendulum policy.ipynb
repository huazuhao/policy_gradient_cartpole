```python
import jsonpickle
```

```python
learned_parameter_theta_file_name = 'learned_parameter.json'
```

```python
learned_parameter_theta_dict = jsonpickle.decode(open(learned_parameter_theta_file_name).read())
```

```python
learned_parameter_theta_dict
```

```python
theta = learned_parameter_theta_dict['theta']
```

```python
theta.shape
```

```python
import gym
import utils
import config as C
import numpy as np

env = gym.make('Pendulum-v0')
```

```python
import policy
policy_outputs_size = env.action_space.shape[0]
policy = policy.Policy(C.extracted_feature_size, policy_outputs_size)
policy.theta = theta

normalizer = utils.Normalizer(C.extracted_feature_size)
normalizer.n = learned_parameter_theta_dict['normalizer_n']
normalizer.mean = learned_parameter_theta_dict['normalizer_mean']
normalizer.mean_diff = learned_parameter_theta_dict['normalizer_mean_diff']
normalizer.var = learned_parameter_theta_dict['normalizer_var']
```

```python
rewards = []

#test the training result
state = env.reset()

while True:

    env.render()
    
    state = utils.extract_features(state)
    normalizer.observe(state)
    state = normalizer.normalize(state)
    
    #compute an action given current observation
    delta = None
    direction = None
    action = policy.evaluate(state, delta, direction)
    
    #apply the action to the environment
    state, reward, done, info = env.step(action)

    rewards.append(reward)
    
    if done:
        env.close()
        break
```

```python
print(np.sum(rewards))
```
