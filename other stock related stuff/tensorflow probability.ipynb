```python
import tensorflow_probability as tfp
tfd = tfp.distributions

a = 0.5
b = 0.5


dist = tfd.Kumaraswamy(a, b)
```

```python
sample = dist.sample(1).numpy()
print(sample[0])
```

```python
sample = dist.sample(1000).numpy()
```

```python
import matplotlib.pyplot as plt

plt.hist(sample,bins = 100)
plt.show()
```

```python
from scipy.stats import uniform

sample = []

for _ in range(0,1000):
    
    #now we sample
    r = uniform.rvs(size = 1)
    action = (1-(1-r)**(1/b))**(1/a) #solve for the cdf of the distribution
    
    sample.append(action[0])
```

```python

plt.hist(sample,bins = 100)
plt.show()
```
